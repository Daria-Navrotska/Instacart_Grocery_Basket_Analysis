# If-Statements with User-Defined Functions

```python
# defining the function 'price_label'

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function to subset

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

# Suppose your clients have changed their minds about the labels you created in your “busiest_day” column.

“Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days.

```python
# create a new column “busiest days” using for-loop
results = []

for value in df_ords_prods_merge["order_dow"]:
  if value in (0, 1):
    results.append("Busiest days")
  elif value in (3, 4):
    results.append("Slowest days")
  else:
    results.append("Regular days")
```

Saturday (0) and Sunday (1) are the busiest days,
Wednesday (3) and Thursday (4) are the Slowest days,

On every loop, the 'if value in (0, 1)' will check if the day is either Saturday (0) OR Sunday (1); if it is, then it will add 'Busiest days' to 'results' list.

If it's not Saturday OR Sunday, the second line (elif value in (3, 4:) will execute and check whether the day is either Wednesday (3) or Thursday (4). Then 'Slowest days' will be printed to the 'results' list.

If none of the two previous conditions is met, then 'else' will be executed, meaning that 'Regular days' will be pushed to the results list.

```python
results
```

```python
# creating the 'busiest_days' column
df_ords_prods_merge['busiest_days'] = results
```

# Identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# receive list from the busiest hour to the least busy hour
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Most orders happened on 9,10,11,12,13,14,15,16 hour of the day; Fewest orders - 0,1,2,3,4,5,6,23 hour of the day; Average orders in the middle - 7, 8, 17-22 hour of the day.

```python
results_hours = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value in (9, 10, 11, 12, 13, 14, 15, 16):
        results_hours.append("Most orders")
    elif value in (7, 8, 17, 18, 19, 20, 21, 22):
        results_hours.append("Average orders")
    else:
        results_hours.append("Fewest orders")
```

```python
# creating the column 'busiest_period_of_day'
df_ords_prods_merge['busiest_period_of_day'] = results_hours
```

```python
# the frequency in the 'busiest_period_of_day'
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```

The sum of the above values is 222406041

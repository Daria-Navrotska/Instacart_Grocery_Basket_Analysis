```python
df_ords_prods_merge_grouped = pd.read_parquet(r'C:\Users\ZenBook\Instacart Basket Analysis_30.09.2025\02 Data\Prepared Data\ords_prods_merge_grouped.parquet')
```

combining will be done by the user_id column as it is common for two databases

```python
#Drop the '_merge' column from the 'ords_prod_merge_grouped'
df_ords_prods_merge_grouped = df_ords_prods_merge_grouped.drop('_merge', axis=1)
```
